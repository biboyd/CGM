```python
import yt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Plotting the Number of Rays over projected area for each bin
want it to be uniform in proj area space so we should get a constant

```python
binned = np.load("impact_par_selection/impact_param.npy")
in_out_rad = np.load("impact_par_selection/in_out_radii.npy")
```

```python
proj_area= np.pi*(in_out_rad[:, 1]**2 - in_out_rad[:, 0]**2)
#look at number of rays per bin
for i in range(len(binned)):
    plt.plot(i, proj_area[i]/len(binned[i]), 'bo')

plt.title("Ray selection with proj area")
plt.xlabel("Bin number")
plt.ylabel("(proj area)/(number of rays)")
```

```python
#just inverse plot of above
for i in range(len(binned)):
    plt.plot(i, len(binned[i]), 'bo')

plt.title("Number of rays per Bin")
plt.xlabel("Bin number")
plt.ylabel("number of rays")
```

# Looking at distribution of Impact Parameter 
in total and in each bin

```python
impact_param=[]

for lis in binned:
    for j in lis:
        impact_param.append(j)
        
```

```python
for i in range(len(impact_param)):
    impact_param[i] = impact_param[i].in_units('kpc')
```

```python
plt.hist(impact_param, bins=25)
plt.title("Distribution of Impact Parameter")
plt.xlabel("impact param in kpc")
plt.ylabel("count")
```

## For each bin

```python
#convert binned to kpc
kpc_bin=[]
for i in range(len(binned)):
    kpc_bin.append([])
    for j in range(len(binned[i])):
        kpc_bin[i].append(binned[i][j].in_units('kpc'))
```

```python
#plot histogram for each
num_bins = len(kpc_bin)
fig=plt.figure(1, figsize=(8, 10))

rows = int(num_bins/2)
cols = 2
for i in range(num_bins):
    
    fig.add_subplot(rows, cols, i+1)
    plt.hist(kpc_bin[i], bins=5)
    plt.title("{} - {} kpc".format(in_out_rad[i, 0], in_out_rad[i, 1]))
    plt.xlabel("Impact parameter (kpc)")
fig.tight_layout()
```
